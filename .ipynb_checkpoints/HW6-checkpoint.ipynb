# Question1

```python
from hdf5_dataset import HDF5Dataset 
import h5py
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sn
from sklearn.metrics import confusion_matrix
```

```python
train_set = HDF5Dataset(file_path = "data/mnist_traindata.hdf5", data_name = 'xdata', label_name = 'ydata')
train_loader = torch.utils.data.DataLoader(train_set, batch_size=100, shuffle=True)

with h5py.File("data/mnist_testdata.hdf5",'r+') as df: 
    test_x=df['xdata'][:] 
    test_y=df['ydata'][:]
```

## The best Test Accurate appeared at epochs=100, lr=0.01, weight_decay=0.0001

#### epochs=100, lr=0.01, without regularization

```python
model=nn.Sequential(nn.Linear(784,10))
criterion=nn.CrossEntropyLoss()
optimizer=optim.SGD(model.parameters(),lr=0.01)
epochs=100

train_loss=[1]
test_loss=[1]
train_acc=[0]
test_acc=[0]
for epoch in range(epochs):
    model.train()
    loss_train=0
    acc_train=0
    for train_x,train_y in train_loader:
        output=model(train_x)
        loss=criterion(output,train_y)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        loss_train+=loss.item()
        #print(np.argmax(output.detach().numpy(),axis=1),np.argmax(train_y,axis=1))
        acc_train+=np.sum(np.argmax(output.detach().numpy(),axis=1)==np.argmax(train_y.detach().numpy(),axis=1))
        #print(acc_train)
    train_loss.append(loss_train/len(train_loader.dataset))
    train_acc.append(acc_train/len(train_loader.dataset))

    with torch.no_grad():
        output=model(torch.tensor(test_x))
        loss=criterion(output,torch.tensor(test_y))
        test_loss.append(loss.item()/len(test_y))
        test_acc.append(np.sum(np.argmax(output.detach().numpy(),axis=1)==np.argmax(test_y,axis=1))/len(test_y))
    print('Epoch:',epoch+1,'Test loss:',test_loss[-1],'Test Accurate:',test_acc[-1])

plt.plot(range(epochs+1),train_loss)
plt.plot(range(epochs+1),test_loss)
plt.legend(['Train Loss','Test Loss'])
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Train & Test Loss')
plt.show()

plt.plot(range(epochs+1),train_acc)
plt.plot(range(epochs+1),test_acc)
plt.legend(['Train Accuracy','Test Accuracy'])
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Train & Test Accuracy')
plt.show()

cf = confusion_matrix(np.argmax(test_y,axis=1),np.argmax(output.detach().numpy(),axis=1))
sn.heatmap(cf,annot=True,fmt='0')      
```

# Question2
### From the two weight histograms below, we can see that the weight distribution in the second model is more spread out compared to the first model. Additionally, the second model has less frequency for the same weight values compared to the first model. This suggests that the regularization technique applied in the second model (L2 regularization and dropout) encourages the network to have a wider range of weights, which can help prevent overfitting by reducing the network's reliance on a few dominant weights.

```python
import torch
import numpy as np
import torchvision
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
```

```python
class Network1(nn.Module):
    def __init__(self):
        super(Network1,self).__init__()
        self.Linear1=nn.Linear(784,128)
        self.ReLU=nn.ReLU()
        self.Linear2=nn.Linear(128,10)
    def forward(self,x):
        x=self.Linear1(x)
        x=self.ReLU(x)
        x=self.Linear2(x)
        return x

class Network2(nn.Module):
    def __init__(self):
        super(Network2,self).__init__()
        self.hiden=nn.Sequential(
            nn.Linear(784,48),
            nn.ReLU(),
            nn.Dropout(0.2)
        )  
        self.Linear=nn.Linear(48,10)
    def forward(self,x):
        x=self.hiden(x)
        x=self.Linear(x)
        return x

def weight_plot (params,mode):
    _,ax = plt.subplots (1, 2, figsize = (10, 10))
    weight_layer1= params[0].cpu().detach() .numpy ()
    ax[0].set_title(f'Input Layer Weights of Network {mode}')
    ax[0].hist (weight_layer1.flatten(), bins = 100, density= True, label='Layer 1')
    ax[0].set_xlabel ('Weights')
    ax[0].set_ylabel ('Frequency')
    ax[0].legend ()
    ### Draw the histogram for layer 2 in a similar way here:
    weight_layer2= params[2].cpu().detach().numpy ()
    ax[1].set_title(f'Hidden Layer Weights of Network {mode}')
    ax[1].hist (weight_layer2.flatten(), bins = 100, density= True, label='Layer 2')
    ax[1].set_xlabel ('Weights')
    ax[1].set_ylabel ('Frequency')
    ax[1].legend ()
    plt.tight_layout()
    plt.show()
```

```python
train_dataset = torchvision.datasets.FashionMNIST(root='./data', train=True, download=True, transform=torchvision.transforms.ToTensor())
test_dataset = torchvision.datasets.FashionMNIST(root='./data', train=False, download=True, transform=torchvision.transforms.ToTensor())
train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=100, shuffle=True)
test_loader = torch.utils.data.DataLoader(test_dataset, batch_size=100)
```

```python
model1=Network1()
criterion1=nn.CrossEntropyLoss()
optimizer1=optim.SGD(model1.parameters(),lr=0.01)
epochs=40

train_loss1=[1]
test_loss1=[1]
train_acc1=[0]
test_acc1=[0]
for epoch in range(epochs):
    model1.train()
    loss_train1=0
    acc_train1=0
    for train_x,train_y in train_loader:
        output1=model1(train_x.view(-1,784))
        loss1=criterion1(output1,train_y)
        optimizer1.zero_grad()
        loss1.backward()
        optimizer1.step()
        loss_train1+=loss1.item()
        acc_train1+=np.sum(np.argmax(output1.detach().numpy(),axis=1)==train_y.detach().numpy())
    train_loss1.append(loss_train1/len(train_loader.dataset))
    train_acc1.append(acc_train1/len(train_loader.dataset))
    print('Epoch:',epoch+1,'Test loss:',train_loss1[-1],'Test Accurate:',train_acc1[-1])
h_params=[params for params in model1.parameters()]
weight_plot(h_params,1)
```

```python
model2=Network2()
criterion2=nn.CrossEntropyLoss()
optimizer2=optim.SGD(model2.parameters(),lr=0.01,weight_decay=0.0001)
epochs=40

train_loss2=[1]
test_loss2=[1]
train_acc2=[0]
test_acc2=[0]
for epoch in range(epochs):
    model2.train()
    loss_train2=0
    acc_train2=0
    for train_x,train_y in train_loader:
        output2=model2(train_x.view(-1,784))
        loss2=criterion2(output2,train_y)
        optimizer2.zero_grad()
        loss2.backward()
        optimizer2.step()
        loss_train2+=loss2.item()
        acc_train2+=np.sum(np.argmax(output2.detach().numpy(),axis=1)==train_y.detach().numpy())
    train_loss2.append(loss_train2/len(train_loader.dataset))
    train_acc2.append(acc_train2/len(train_loader.dataset))
    print('Epoch:',epoch+1,'Test loss:',train_loss2[-1],'Test Accurate:',train_acc2[-1])
h_params=[params for params in model2.parameters()]
weight_plot(h_params,2)
```

# Question3

```python
import torch
import numpy as np
import torchvision
import seaborn as sn
import torch.nn as nn
import matplotlib.pyplot as plt
import torch.optim as optim
import torchvision.transforms as transforms
from sklearn.metrics import confusion_matrix
```

```python
class MLP(nn.Module):
    def __init__(self):
        super(MLP,self).__init__()
        self.hiden1=nn.Sequential(
            nn.Linear(32*32*3,256),
            nn.ReLU(),
            nn.Dropout(0.3)
        )
        self.hiden2=nn.Sequential(
            nn.Linear(256,128),
            nn.ReLU(),
            nn.Dropout(0.3)
        )
        self.Linear=nn.Linear(128,10)
    def forward(self,x):
        x=self.hiden1(x)
        x=self.hiden2(x)
        x=self.Linear(x)
        return x
    
def dataloader(mean,std,batch_size):
    transform=transforms.Compose([transforms.ToTensor(),transforms.Normalize(mean,std)])
    train_dataset = torchvision.datasets.CIFAR10(root='./data', train=True, download=True, transform=transform)
    test_dataset = torchvision.datasets.CIFAR10(root='./data', train=False, download=True, transform=transform)
    train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    test_loader = torch.utils.data.DataLoader(test_dataset, batch_size=batch_size)
    classes=train_dataset.classes
    return train_loader,test_loader,classes
    
    
mean = np.array((0.4914,0.4822,0.4465),dtype=np.float32)
std = np. array ((0.247,0.243,0.261), dtype=np.float32)
batch_size=100
train,test,classes=dataloader(mean,std,batch_size)
```

```python
model3=MLP()
criterion3=nn.CrossEntropyLoss()
optimizer3=optim.SGD(model3.parameters(),lr=0.01,weight_decay=0.0001)
epochs=50

train_loss3=[1]
test_loss3=[1]
train_acc3=[0]
test_acc3=[0]
Out=[]
test_target=[]
for epoch in range(epochs):
    model3.train()
    loss_train3=0
    acc_train3=0
    loss_test3=0
    acc_test3=0
    for train_x,train_y in train:
        output3=model3(train_x.view(-1,32*32*3))
        loss3=criterion3(output3,train_y)
        optimizer3.zero_grad()
        loss3.backward()
        optimizer3.step()
        loss_train3+=loss3.item()
        acc_train3+=np.sum(np.argmax(output3.detach().numpy(),axis=1)==train_y.detach().numpy())
    train_loss3.append(loss_train3/len(train.dataset))
    train_acc3.append(acc_train3/len(train.dataset))
    with torch.no_grad():
        for test_x, test_y in test:
            output3=model3(test_x.view(-1,32*32*3))
            loss3=criterion3(output3,test_y)
            loss_test3+=loss3.item()
            acc_test3+=np.sum(np.argmax(output3.detach().numpy(),axis=1)==test_y.detach().numpy())
            '''Out=np.concatenate((Out,np.argmax(output3.detach().numpy(),axis=1)),axis=None)
            test_target.append(test_y.detach().numpy())'''
        test_loss3.append(loss_test3/len(test.dataset))
        test_acc3.append(acc_test3/len(test.dataset))
    print('Epoch:',epoch+1,'Test loss:',test_loss3[-1],'Test Accurate:',test_acc3[-1])

for test_x, test_y in test:
    output3=model3(test_x.view(-1,32*32*3))
    acc_test3+=np.sum(np.argmax(output3.detach().numpy(),axis=1)==test_y.detach().numpy())
    Out=np.concatenate((Out,np.argmax(output3.detach().numpy(),axis=1)),axis=None)
    test_target=np.concatenate((test_target,test_y.detach().numpy()),axis=None)

cf = confusion_matrix(test_target,Out)
fig, ax = plt.subplots(figsize=(10,10)) 
ax=sn.heatmap(cf,annot=True,fmt='0',ax=ax)
ax.xaxis.set_ticklabels(classes,fontsize=9)
ax.yaxis.set_ticklabels(classes,fontsize=9)
plt.show()
```
